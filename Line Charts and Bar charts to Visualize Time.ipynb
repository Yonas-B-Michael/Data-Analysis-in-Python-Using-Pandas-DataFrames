# Lets answer to some analytical questions 

```python
question 1 - In which year the highest and lowest number of movies?

Now finding out number of movies per year 
```

```python
import pandas as pd
```

```python
data= pd.read_json("movies_dataset.json")
```

```python
data.head()
```

Now finding out number of movies per year

```python
count_per_year=data.groupby("title_year").size()
count_per_year.max()
count_per_year.idxmax()
```

lets create a line chart to show the result

```python
ax=count_per_year.plot(figsize=(8,6))
ax.set_xlabel("years")
ax.set_ylabel("movies")
```

Question-2 - In which year we have minimum and maximum average imdb_score?
        

```python
avgscores= data.groupby("title_year")["imdb_score"].mean()
avgscores
```

Lets use a barchart to visualize the result 

```python
ax = avgscores.plot.bar(figsize=(18,8), title="Average Imdb Score year")
ax.set_xlabel("title year")
ax.set_ylabel("avg score")
```

Question 3 - In which year we have the minimum and maximum spent movie budget?

```python
data.groupby("title_year")["budget"].sum().plot()
```
